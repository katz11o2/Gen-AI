```python
import gradio as gr
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch
from diffusers import StableDiffusionPipeline
from transformers import BlipProcessor, BlipForConditionalGeneration


model_name = "gpt2"
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

def generate_caption(prompt):
    try:
        inputs = tokenizer.encode(prompt, return_tensors="pt")
        outputs = model.generate(
            inputs,
            max_length=500,
            num_return_sequences=1,
            no_repeat_ngram_size=2,
            top_p=0.95,
            top_k=60,
            temperature=0.7,
            pad_token_id=tokenizer.eos_token_id
        )
        caption = tokenizer.decode(outputs[0], skip_special_tokens=True)
        words = caption.split()
        if len(words) > 400:
            caption = ' '.join(words[:500])
        return caption
    except Exception as e:
        return f"Error: {str(e)}"


model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe.to("cuda")

def generate_image(prompt: str):
    image = pipe(prompt).images[0]
    return image


processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model_blip = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")

def generate_image_caption(image):
    inputs = processor(images=image, return_tensors="pt")
    out = model_blip.generate(**inputs)
    caption = processor.decode(out[0], skip_special_tokens=True)
    return caption


def gradio_app():
    with gr.Blocks() as demo:
        gr.Markdown("# Multi-Function Caption and Image Generator")


        with gr.Tab("Generate Post Caption"):
            prompt_input = gr.Textbox(label="Enter your post prompt here", placeholder="Type your idea or description...")
            caption_output = gr.Textbox(label="Generated Caption")
            submit_button_1 = gr.Button("Submit")
            submit_button_1.click(generate_caption, inputs=prompt_input, outputs=caption_output)


        with gr.Tab("Generate Image from Text"):
            text_input = gr.Textbox(label="Enter a text prompt", placeholder="Describe an image you want to generate...")
            image_output = gr.Image(label="Generated Image")
            submit_button_2 = gr.Button("Submit")
            submit_button_2.click(generate_image, inputs=text_input, outputs=image_output)


        with gr.Tab("Generate Caption from Image"):
            image_input = gr.Image(type="pil", label="Upload an image")
            image_caption_output = gr.Textbox(label="Generated Caption")
            submit_button_3 = gr.Button("Submit")
            submit_button_3.click(generate_image_caption, inputs=image_input, outputs=image_caption_output)

    demo.launch()


gradio_app()
```
